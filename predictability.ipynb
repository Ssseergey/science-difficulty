```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
highlab = ['survival', 'socialization', 'self_realization']
```

```python
hor = pd.read_csv('results/res_test.csv')
hor.round(2)
```

```python
hor['aggregated'] = np.sqrt((hor[highlab] ** 2).sum(axis=1) / 3.)
hor.id = hor.id.astype(str)
hor.describe()
```

```python
hor.hist(bins=20)
```

```python
x = np.linspace(-0., 1., 25)
columns = ['survival', 'socialization', 'self_realization', 'aggregated']
colors = ['blue', 'green', 'darkorange', 'magenta']
labels = ['Hit probability (F1>3/4)'] * 3 + ['Aggregated predictability']
plt.figure(figsize=(9, 7))
for i, j in enumerate(zip(columns, colors, labels)):
    plt.subplot(2, 2, i + 1)
    plt.title('For ' + j[0], size=14)
    plt.hist(hor[j[0]], x, width=0.03, color=j[1], align='mid')
    plt.xlabel(j[2], size=14)
    if not i % 2:
        plt.ylabel('Number of customers', size=14)
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-0., 1., 10)
for i, v in enumerate(highlab):
    plt.hist(hor[v] + 0.1, x + i / 33.3, width=0.03, label=v.capitalize(), align='mid')
for i in x[:]:
    plt.axvline(i - 0.003, ls='--', c='gray')
plt.legend(fontsize=14)
plt.xticks(x[:]-.005, (x[:]).round(1))
plt.xlabel('Hit-probability', size=14)
plt.ylabel('Number of customers', size=14)
plt.show()
```

```python
compression = pd.read_csv("processed/transact_18_22_lempel_ziv_compression.csv", index_col="client")
compression.head(5)
```

```python
columns = ['survival', 'socialization', 'self_realization']
colors = ['blue', 'green', 'darkorange']
labels = ['Hit probability (F1 > 0.75)'] * 3

plt.figure(figsize=(9, 7))

for i, (col, color, label) in enumerate(zip(columns, colors, labels)):
    plt.subplot(2, 2, i + 1)

    plt.scatter(
        compression[col],
        hor[col],
        color=color,
        alpha=0.6,
        s=25
    )

    plt.title(f'For {col}', size=14)
    plt.xlabel('LZ-complexity', size=14)

    if i % 2 == 0:
        plt.ylabel(label, size=14)

plt.tight_layout()
plt.show()
```
